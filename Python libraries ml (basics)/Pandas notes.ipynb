# Series

```python
student=pd.Series([1,2,3,4]) # indices are set default.
print(student)
```

```python
stu=pd.DataFrame({'columnA': student,'columnB':pd.Series([5,6,7,8]) }) # Here two series are used to create a Dataframe.
stu
```

Adding two series using '+' sign add the corresponding values in each series . If there is any null value in series , the whole datatype turns into float64.

```python
s1= pd.Series([1,2,3,8])+ pd.Series([4,5,6,7])
s1.astype(int)
```

```python
s2 = pd.Series([1,2,3],index = ['one','two','three']).apply(lambda x :x**2)#apply-applies the function to each element of series
s2
```

```python
s2=s2.map({1:'0',4:'2'}) # Here map() is used to replace the elements with given values.
s2
```

```python
s3 = pd.Series(['great  ','good','awesome'])
s3.map('I am {}'.format)
```

```python
s3.describe()
```

```python
s4 = pd.Series([1,2,2,3,3,3,4,4,5,6])
[s4>3] # shows the boolean value according to the condition given.
```

```python
s4.value_counts() #shows frequency of each element in the series.
```

```python
s1[s1>3] #In this format, the condition applies to both indices and values of the series
```

Merging eat and eat1 dataframes.

```python
eat.merge(eat1) # Here, the rows with unmatched values are removed.
```

```python
eat1.merge(eat) 
```

```python
eat.merge(eat1,how = 'outer') # outer - union of both the dataframes , default is inner merge.
```

```python
eat.merge(eat1,how = 'left') # shows all the values from left i.e first dataframe.
```

```python
eat.merge(eat1,how = 'right') # right - shows all the values from right i.e second dataframe.
```

```python
eat1.merge(eat,how = 'inner',on = ['bananas','Names'])
```

```python
eat1.merge(eat,how = 'outer',on = ['bananas','Names'])
```

```python
eat1.merge(eat,how = 'outer',on = ['bananas','Names','oramges'])
```
