# Various plots in Matplotlib

```python
from matplotlib import pyplot as plt
```

# Simple plots

```python
plt.plot([1,2,3],[1,2,3.5])
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.title("square")
plt.ylabel('y axis')
plt.xlabel("x axis")
plt.grid(True,color ='g')
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('bmh')
x2=[1,2,3]
y2=[16,25,36]
plt.plot(x,y,'blue',label = 'first',linewidth=3)
plt.plot(x2,y2,'black',label = 'second',linewidth=3)
plt.title("square")
plt.ylabel('y axis')
plt.xlabel("x axis")

plt.legend()
plt.grid(True,color ='g')
plt.show()
```

# Bargraph

```python
plt.bar(x,y,label = 'first',color = 'blue')
plt.bar(x2,y2,label = 'second',color = 'green')
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar graph')
plt.legend()
plt.show()
```

# Histogram

```python
plt.hist(x,y,histtype='bar',rwidth=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
plt.show()
```

# Scatter plot

```python
x1=[1,2,3,4,5,6,4,0,7,9,7]
y1=[1,6,3,4,3,3,3,5,7,2,5]
plt.scatter(x1,y1,label="Cluster")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

# Area plot

```python
days = [1,2,3,4,5,6]
rahul=[0,2,3,4,1,5]
sowmya=[8,4,3,5,6,7]
plt.plot([],[],color = 'm',label = 'rahul')
plt.plot([],[],color = 'r',label= 'sowmya')
plt.stackplot(days,rahul,sowmya,colors=['m','r'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area plot')
plt.legend()
plt.show()
```

# Pie chart

```python
#no need of formig lists of rahul,sowmya,someone 
s=[3,5,4]
activities=['rahul','sowmya','someone']
cols = ['m','r','c']
plt.pie( s,labels = activities , colors = cols , startangle = 180 , explode  = (0.1,0,0) , autopct = '%1.1f%%')
plt.title('Pie chart')
plt.show()
```

# Multiple plots

```python
import numpy as np
```

```python
def fun(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,6.0,0.03)
plt.subplot(211)
plt.plot(t1,fun(t1),'bo')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
