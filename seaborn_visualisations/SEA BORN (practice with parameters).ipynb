```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# BAR PLOTS

### Impementing with different parameters and features

```python
tips = sb.load_dataset('tips')
```

```python
sb.barplot(x = 'day', y = 'tip', data = tips )
```

```python
sb.barplot(x = 'day', y = 'tip', data = tips, hue = 'sex' )
```

```python
sb.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'spring' )
```

```python
sb.barplot(x = 'day', y = 'total_bill', data = tips , hue =  'smoker', palette = 'winter_r' )
```

```python
sb.barplot(x = 'sex', y = 'tip', data = tips, palette = 'husl', ci = 99 )
```

```python
sb.barplot(x = 'day', y = 'tip', data = tips, capsize = 0.3,palette = 'autumn')
```

```python
sb.barplot(x = 'smoker', y = 'tip', data = tips, estimator = np.mean , palette = 'coolwarm')
```

```python
sb.barplot(x = 'day', y = 'tip', data = tips, estimator = np.mean )
```

```python
sb.barplot(x = 'size', y = 'tip', data = tips)
```

# DIST PLOTS

```python
num = np.random.randn(150)
```

```python
sb.distplot(num)
```

```python
sb.distplot(num, color = 'red')
```

```python
label = pd.Series(num, name = 'varaible x')
```

```python
sb.distplot(label)
```

```python
sb.distplot(label, vertical = 'True')
```

```python
sb.distplot(label , hist = False)
```

```python
sb.distplot(label , hist = False, rug = True)
```

```python
sb.distplot(label , rug = True , color = 'green')
```
