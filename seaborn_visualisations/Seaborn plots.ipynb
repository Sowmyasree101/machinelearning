# Data visualisations through seaborn library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
flight_data= sb.load_dataset("flights")
flight_data.head()
```

```python
sb.relplot(x="passengers",y="month",hue = "year",data = flight_data)
```

```python
tips_data=sb.load_dataset("tips")
tips_data.head()
```

### SCATTER PLOT

```python
sb.scatterplot(data = tips_data,x = 'total_bill',y='tip')
#scatter plot shows the realtion between two variables (how one is affected by other) and is useful in finding thier correlation.
```

```python
sb.relplot(x="time",y="total_bill",data = tips_data,kind="line")
```

### CAT PLOTS

```python
sb.catplot(x="day",y="total_bill",data = tips_data)
# This plot shows relation between numerical and one or more categorical features and explains its distribution over these features.
```

```python
sb.catplot(x="day",y="total_bill",data = tips_data,kind="boxen") 
# catplot has access to several types of plots like boxen,violin etc
```

```python
sb.set(style="whitegrid",color_codes=False)
sb.catplot(x="day",y="total_bill",data = tips_data,kind="boxen")
```

```python
sb.catplot(x="day",y="total_bill",data = tips_data,kind="violin")
```

### DIST PLOT

```python
data2 = np.random.normal(loc=5,size=100,scale = 2)
sb.distplot(data2)
# Distplot shows the univariate distribution of variable against denisty distribution.
```

### JOINT PLOT

```python
x=pd.DataFrame({"Day":[1,2,3,4,5],"Grocery":[3,4,2,6,1],"clothes":[4,5,6,8,6]})
y=pd.DataFrame({"Day":[5,6,7,8,9],"Grocery":[2,5,4,7,8],"clothes":[7,6,10,8,9]})
mean, cov= [0,1],[(1,.5),(.5,1)]
data = np.random.multivariate_normal(mean,cov,200)
with sb.axes_style("white"):
    sb.jointplot(x=x,y=y,kind="kde",color="b"); #kde is kernel density estimator
# This is used to show relation between two variables and describe thier individual distributions on the same plot
```

```python
iris_data = sb.load_dataset("iris")
iris_data.head()
```

### FACET GRID

```python
fig=sb.FacetGrid(iris_data,col="species")
fig.map(plt.hist,"sepal_length")
```

### BOX PLOT

```python
sb.set(style="white",palette = 'spring')
sb.boxplot(x="day",y="total_bill",data=tips_data)
```

```python
sb.set(style="white",palette = 'winter_r')
sb.boxplot(x="day",y="total_bill",data=tips_data)
sb.despine(offset=10,trim=True)
```
